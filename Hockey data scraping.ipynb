```python
import csv
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
# definisco url
url = 'https://www.scrapethissite.com/pages/forms/?per_page=50'
```

```python
# effettuo la richiesta
page = requests.get(url)
```

```python
# stampo lo stato della response [200=ok]
print(page)
```

```python
#print(page.text)
```

```python
# utilizzo BeautifulSoup to analyze html
soup = BeautifulSoup(page.text, 'html')
```

```python
# trovo la tabella che mi interessa indicando elemento e classe
table = soup.find('table', class_='table')
print(table)
```

```python
# column name to extract
table.find_all('th')
```

```python
# cerco di ottenere una lista dei nomi delle colonne
column_names = table.find_all('th')
column_names_list = [name.text.strip() for name in column_names]
print(column_names_list)
```

```python
# cerco di ottenere una lista delle righe individuali della tabella
data_rows = table.find_all('tr', class_='team')

for row in data_rows:
  # estraggo tutti i td dalla riga corrente
  cells = row.find_all('td')
  # estraggo il testo
  individual_row = [cell.text.strip() for cell in cells]
  print(individual_row)
```

```python
# memorizzo la lista di righe
all_rows = []
data_rows = table.find_all('tr', class_='team')

for row in data_rows:
    cells = row.find_all('td')
    individual_row = [cell.text.strip() for cell in cells]
    # aggiungo riga attuale alla lista
    all_rows.append(individual_row)

print(all_rows)
```

```python
# salvo tutto in un csv

filename = 'hockey_DataScraping.csv'

with open(filename, 'w', newline='') as file:
  # oggetto writer
  writer = csv.writer(file)
  # header
  writer.writerow(column_names_list)
  # rows
  writer.writerows(all_rows)

print(f'File {filename} scritto con successo')
```
